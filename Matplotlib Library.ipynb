# Bar Charts

```python
x=[2,8,10]
y=[11,15,9]

x2=[12,5,15]
y2=[6,8,10]
plt.bar(x,y,color='m')
plt.bar(x2,y2, color='g')
plt.title("Bar plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.figure(figsize=(8,8))  
plt.bar([1, 3, 5, 7, 9], [5, 2, 7, 8, 2], label = 'Instance 1', color='g')
plt.bar([2, 4, 6, 8, 10], [8, 6, 2, 5, 6], label = 'Instance 2', color ='r')

plt.legend()

plt.ylabel('Bar number')
plt.xlabel('Bar height')

plt.title("Bar plot")

plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
plt.bar(company,revenue,label='revenue', color='b')
plt.legend()
plt.xlabel("Stocks")
plt.ylabel("revenue");
```

```python
#plotting revenue and profit side by side
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]


plt.bar(company,revenue,width=0.3,label='revenue', color='r')
plt.bar(company,profit,color='blue',label='profit')
plt.legend()
plt.xlabel("Stocks")
plt.ylabel("revenue and profit");
```

# Scatter plot

```python
plt.figure(figsize=(5,5))
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

x1=[3,8,4]
y1=[6,4,9]
plt.scatter(x, y, label = 'cat', color = 'k', s=60 , marker = "^")  # s is marker size
plt.scatter(x1, y1, label='weight',color='blue', s=100, marker="+")  

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")

plt.legend(loc='lower right')
plt.show()
```

# Stack Plots

```python
days = [1, 2, 3, 4, 5]

sleeping = [7, 8, 6, 11, 7]
eating = [2, 3, 4, 3, 2]
coding = [7, 8, 7, 2, 2]
playing = [8, 5, 7, 8, 13]

plt.plot([], [], color = 'b', label = 'Sleeping', linewidth = 5)
plt.plot([], [], color = 'g', label = 'eating', linewidth = 5)
plt.plot([], [], color = 'c', label = 'coding', linewidth = 5)
plt.plot([], [], color = 'm', label = 'playing', linewidth = 5)

plt.stackplot(days, sleeping, eating, coding, playing, colors = ['b', 'g', 'c', 'm'])


plt.xlabel("Days")
plt.ylabel("Sleeping. eating,coding,playing")
plt.title("Stack Plot")

plt.legend()
plt.show()
```

# Pie Charts

```python
fractions = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'coding', 'playing']
cols = ['c', 'm', 'r', 'b']

plt.pie(fractions , labels = activities, colors = cols, startangle=90, shadow = True, explode = (0, 0.5, 0, 0), autopct = '%1.1f%%')

plt.title('Pie Plot')
plt.show()
```
